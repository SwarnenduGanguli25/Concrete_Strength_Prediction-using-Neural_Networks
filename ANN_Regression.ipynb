##### Splitting the data into predictors column and target column

```python
dataset_columns = dataset.columns

predictors = dataset[dataset_columns[dataset_columns != 'Strength']] # all columns except Strength
target = dataset['Strength'] # Strength column
```

##### Splitting into train and test dataset

```python
from sklearn.model_selection import train_test_split
```

```python
predictors_train, predictors_test, target_train, target_test = train_test_split(predictors, target, test_size=0.3, random_state=0)
```

```python
predictors_train.head()
```

```python
predictors_train.shape
```

```python
predictors_test.head()
```

```python
predictors_test.shape
```

#### Now creating a neural network

```python
n_cols = predictors_train.shape[1] # number of predictors columns
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
# define regression model
def regression_model():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

THIS WILL CREATE 1 HIDDEN LAYER OF 10 NODES

#### Training and testing the network

```python
# build the model
model = regression_model()
```

```python
# fit the model
model.fit(predictors_train, target_train, validation_data=(predictors_test, target_test), epochs=50, verbose=2)
```

```python
predictions = model.predict(predictors_test)
```

##### Calculating the mean squared error

```python
from sklearn.metrics import mean_squared_error
```

#### Creating a function for 50 mean squared errors

```python
mse = [] # list for holding all errors
```

```python
def ann_mse():
    predictors_train, predictors_test, target_train, target_test = train_test_split(predictors, target, test_size=0.3, random_state=0)
    model.fit(predictors_train, target_train, validation_data=(predictors_test, target_test), epochs=50, verbose=2)
    predictions = model.predict(predictors_test)
    predictions = predictions.flatten()
    error = mean_squared_error(target_test, predictions)
    mse.append(error)
```

```python
for i in range(50):
    ann_mse()
```

```python
len(mse)
```

#### Calculating the final mean and standard deviation of 50 mean squared errors

```python
mse_array = np.array(mse)
```

```python
print("THE MEAN IS : ",mse_array.mean())
print("THE STANDARD DEVIATION IS : ",mse_array.std())
```

##### Creating a new neural network

```python
n_cols = predictors_norm.shape[1] # number of predictors columns
```

```python
# define regression model
def regression_model_2():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
# Building model
model = regression_model_2()
```
